```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
data = pd.read_csv("Real estate.csv")
data.head()
```

```python
X = data['X3 distance to the nearest MRT station'].values
Y = data['Y house price of unit area'].values
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```python
s1 = 0
s2 = 0

for i in range(414):
  s1 = s1 + (X[i] - mean_x) * (Y[i] - mean_y)
  s2 = s2 + (X[i] - mean_x) ** 2
m = s1/s2
c = mean_y - (m*mean_x)
```

```python
max_x = np.max(X)
min_x = np.min(X)
x = np.linspace(min_x,max_x)
y = m*x+c

plt.plot(x,y,color='red')
plt.scatter(X,Y)
plt.xlabel('X3 distance to the nearest MRT station')
plt.ylabel('Y house price of unit area')
plt.show()
```

```python
suma1 = 0
for i in range(414):
  y = c + m*X[i]
  suma1 = suma1 + ((Y[i] - y) ** 2)**(1/2)
  MAE = (1/414)*suma1
  print(MAE)
```

```python
suma = 0
for i in range(414):
  y =  c + m*X[i]
  suma = suma + ((Y[i] - y) ** 2)/(i+1)
  RMSE = suma**(1/2)
  print(RMSE)
```
